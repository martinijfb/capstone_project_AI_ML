# Data Overview — All Functions

Run this notebook to see the current state of all data across all functions and weeks.

```python
import numpy as np
import pandas as pd

INITIAL_COUNTS = {1: 10, 2: 10, 3: 15, 4: 30, 5: 20, 6: 20, 7: 30, 8: 40}
DIMS = {1: 2, 2: 2, 3: 3, 4: 4, 5: 4, 6: 5, 7: 6, 8: 8}

# --- Summary table ---
rows = []
for n in range(1, 9):
    X = np.load(f'../data/function_{n}/initial_inputs.npy')
    Y = np.load(f'../data/function_{n}/initial_outputs.npy')
    extra = len(Y) - INITIAL_COUNTS[n]
    best_idx = np.argmax(Y)
    rows.append({
        'Function': f'F{n}',
        'Dims': f'{DIMS[n]}D',
        'Total Pts': len(Y),
        'Initial': INITIAL_COUNTS[n],
        'Submissions': extra,
        'Best Y': Y.max(),
        'Best Pt': best_idx,
        'Worst Y': Y.min(),
        'Mean Y': Y.mean(),
    })

df_summary = pd.DataFrame(rows)
print("=" * 80)
print("FUNCTION SUMMARY")
print("=" * 80)
display(df_summary.style.format({
    'Best Y': '{:.4f}', 'Worst Y': '{:.4f}', 'Mean Y': '{:.4f}'
}).set_properties(**{'text-align': 'right'}))
```

```python
# --- Detailed data per function ---
# Shows all points with week labels (initial vs submitted)

for n in range(1, 9):
    X = np.load(f'../data/function_{n}/initial_inputs.npy')
    Y = np.load(f'../data/function_{n}/initial_outputs.npy')
    
    n_initial = INITIAL_COUNTS[n]
    dim_names = [f'x{d+1}' for d in range(X.shape[1])]
    
    # Build dataframe
    df = pd.DataFrame(X, columns=dim_names)
    df['Y'] = Y
    df['Week'] = ['initial'] * n_initial + [f'w{w+1}' for w in range(len(Y) - n_initial)]
    df['Point'] = [f'pt{i}' for i in range(len(Y))]
    df = df[['Point', 'Week'] + dim_names + ['Y']]
    
    # Highlight best
    best_idx = np.argmax(Y)
    
    print(f"\n{'='*80}")
    print(f"FUNCTION {n} ({X.shape[1]}D) — {len(Y)} points, best Y = {Y.max():.4f} (pt{best_idx})")
    print(f"{'='*80}")
    
    # Format and display
    formatters = {col: '{:.4f}' for col in dim_names + ['Y']}
    display(df.style.format(formatters).apply(
        lambda row: ['background-color: #d4edda' if row.name == best_idx else '' for _ in row],
        axis=1
    ))
```

```python
# --- Submission history: only the submitted points across weeks ---

rows = []
for n in range(1, 9):
    X = np.load(f'../data/function_{n}/initial_inputs.npy')
    Y = np.load(f'../data/function_{n}/initial_outputs.npy')
    n_initial = INITIAL_COUNTS[n]
    
    for w in range(len(Y) - n_initial):
        idx = n_initial + w
        x_str = '-'.join(f'{X[idx,d]:.6f}' for d in range(X.shape[1]))
        rows.append({
            'Function': f'F{n}',
            'Week': w + 1,
            'Submitted X': x_str,
            'Returned Y': Y[idx],
            'Was Best at Time?': Y[idx] >= Y[:idx].max() if idx > 0 else True,
        })

if rows:
    df_hist = pd.DataFrame(rows)
    print("\n" + "=" * 80)
    print("SUBMISSION HISTORY")
    print("=" * 80)
    display(df_hist.style.format({'Returned Y': '{:.4f}'}).apply(
        lambda row: ['background-color: #d4edda' if row['Was Best at Time?'] else '' for _ in row],
        axis=1
    ))
else:
    print("No submissions yet.")
```
